```python
from googleapiclient.discovery import build
import pandas as pd

def get_channel_playlists(api_key, channel_id):
    
    youtube = build('youtube', 'v3', developerKey=api_key)

    playlists = []
    next_page_token = None

    while True:
        request = youtube.playlists().list(
            part='snippet',
            channelId=channel_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response['items']:
            playlist_title = item['snippet']['title']
            playlist_id = item['id']
            playlists.append({'title': playlist_title, 'id': playlist_id})

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return playlists

api_key = 'AIzaSyCebl_pUiODay2-vKZhLyi6XNdOYeOzDuu'
channel_id = 'UCbNaKxi1j-Q5jWZDpJZp56g'

playlists = get_channel_playlists(api_key, channel_id)
playlists = pd.DataFrame(playlists)

pd.set_option('display.max_rows', None)
playlists
```

```python
def get_playlist_videos_info(api_key, playlist_id):

    youtube = build('youtube', 'v3', developerKey=api_key)

    videos_info = []
    next_page_token = None

    while True:
        playlist_request = youtube.playlistItems().list(
            part='snippet',
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )
        playlist_response = playlist_request.execute()

        video_ids = [item['snippet']['resourceId']['videoId'] for item in playlist_response['items']]

        video_request = youtube.videos().list(
            part='snippet,contentDetails,statistics',
            id=','.join(video_ids)
        )
        video_response = video_request.execute()

        for item in video_response['items']:
            video_info = {
                'Назва': item['snippet']['title'],
                'ID': item['id'],
                'Опубліковано': item['snippet']['publishedAt'],
                'Тривалість': item['contentDetails']['duration'],
                'Перегляди': item['statistics'].get('viewCount', 0),
                'Likes': item['statistics'].get('likeCount', 0),
                'Dislikes': item['statistics'].get('dislikeCount', 0),
                'Коментарі': item['statistics'].get('commentCount', 0)
            }
            videos_info.append(video_info)

        next_page_token = playlist_response.get('nextPageToken')
        if not next_page_token:
            break

    return videos_info

api_key = 'AIzaSyCebl_pUiODay2-vKZhLyi6XNdOYeOzDuu'
playlist_id = 'PL7joDR5Xn9vetumm4k7jBFXE1tvbbXrQB'

videos_info = get_playlist_videos_info(api_key, playlist_id)
videos_info = pd.DataFrame(videos_info)

pd.set_option('display.max_rows', None)
videos_info
```
