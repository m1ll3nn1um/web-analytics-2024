```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def Page_Parser(allNewsOnPage, L):
    for line in allNewsOnPage:
        link_on_news = line.findAll('a', href=True)
        link_on_news = 'https://mignews.ua' + str(link_on_news[2]['href'])

        Text = line.find('div', class_= "post-content mt-1").text
        Text = "".join([s for s in Text.strip().splitlines(True) if s.strip()])
        Text = Text.split('\n')
        Text[3] = Text[3].lstrip()

        TimeOfPublication = Text[0]
        Rubric = Text[1]
        Title = Text[2]
        Content = Text[3]

        d = {
            'Заголовок': Title,
            'Дата публікації': TimeOfPublication,
            'Вміст': Content,
            'Рубрика': Rubric,
            'Посилання на статтю': link_on_news
        }

        L.append(d)
```

```python
L = []

for i in range(1,61):
    url = 'https://mignews.ua/uk/news/ukraine/?page={i}'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    allNewsOnPage = soup.findAll('div', class_="col-md-6 mb-4")
    Page_Parser(allNewsOnPage, L)
print('* List is formed.')
```

```python
df = pd.DataFrame(L)
df
```

```python
df.to_csv('News.csv', index=False, encoding='utf-8')
```
